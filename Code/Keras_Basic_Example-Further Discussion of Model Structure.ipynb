Solving the question using a one-hidden layer neural network 
y = 6x : 
x = -1 , y = -6 x = -2 , y = -12 ...

Structure :

                    (Linear Activation func.)      (Linear Activation func.)
            
                 w12        Bias12        w23        Bias23
       (x1) --------------> (x2) -----------------> (y3)

```python
import keras
from keras import models
```

```python
model = models.Sequential()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,input_dim=1, activation = 'linear'))
```

```python
model.add(Dense(1,input_dim=1, activation = 'linear'))
```

```python
model.summary()
```

```python
import numpy as np
```

Assume we only know the following input and output 
x = -1 , y = -6
x = -2 , y = -12
x = -4 , y = -24
x = -8  , y = -48
x = -10 , y = -60
x = 1 , y = 6
x = 2 , y = 12
x = 4 , y = 24
x = 8  , y = 48
x = 10 , y = 60


```python
x = np.array([[-1],[-2],[-4],[-8],[-10],[1],[2],[4],[8],[10]])
```

```python
y = np.array([[-6],[-12],[-24],[-48],[-60],[6],[12],[24],[48],[60]])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'ro')
plt.show()
```

The details of the optimizer

```python
#sgd = keras.optimizers.SGD(lr=0.0000001, momentum=0.0, decay=0.0, nesterov=False)
```

```python
sgd = keras.optimizers.SGD(lr=0.000001, nesterov=True) # one of the learning rate in this situation --> lr too large --> loss too large
```

```python
#model.compile(loss='mse', optimizer= 'adam' ,metrics=['acc','mse'])
model.compile(loss='mse', optimizer= sgd ,metrics=['acc','mse'])
```

```python
history = model.fit(x, y , epochs=3000 , verbose=2, shuffle=True)
```

```python
print(model.layers[0].get_weights()[0]) # weights 
print(model.layers[0].get_weights()[1]) # bias
```

```python
print(model.layers[1].get_weights()[0]) # weights 
print(model.layers[1].get_weights()[1]) # bias
```
